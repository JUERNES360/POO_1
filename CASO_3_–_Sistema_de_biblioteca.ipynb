## Caso 3: Sistema de Biblioteca

```python
class MaterialBiblioteca:
    def __init__(self, titulo, autor_o_editorial, anio_publicacion):
        self.titulo = titulo
        self.autor_o_editorial = autor_o_editorial
        self.anio_publicacion = anio_publicacion
        self.disponible = True

    def prestar(self):
        if self.disponible:
            self.disponible = False
            print(f"'{self.titulo}' ha sido prestado.")
        else:
            print(f"'{self.titulo}' no está disponible para préstamo.")

    def devolver(self):
        if not self.disponible:
            self.disponible = True
            print(f"'{self.titulo}' ha sido devuelto.")
        else:
            print(f"'{self.titulo}' ya está en la biblioteca.")

    def __str__(self):
        estado = "Disponible" if self.disponible else "No Disponible"
        return f"Título: {self.titulo}, Autor/Editorial: {self.autor_o_editorial}, Año: {self.anio_publicacion}, Estado: {estado}"


class Libro(MaterialBiblioteca):
    def __init__(self, titulo, autor, anio_publicacion, isbn):
        super().__init__(titulo, autor, anio_publicacion)
        self.isbn = isbn

    def __str__(self):
        base_str = super().__str__()
        return f"Libro - {base_str}, ISBN: {self.isbn}"


class Revista(MaterialBiblioteca):
    def __init__(self, titulo, editorial, anio_publicacion, numero_revista):
        super().__init__(titulo, editorial, anio_publicacion)
        self.numero_revista = numero_revista

    def __str__(self):
        base_str = super().__str__()
        return f"Revista - {base_str}, Número: {self.numero_revista}"


class Tesis(MaterialBiblioteca):
    def __init__(self, titulo, autor, anio_publicacion, universidad):
        super().__init__(titulo, autor, anio_publicacion)
        self.universidad = universidad

    def __str__(self):
        base_str = super().__str__()
        return f"Tesis - {base_str}, Universidad: {self.universidad}"
```

### Demostración de las clases

```python
# Crear instancias de los materiales
libro1 = Libro("Cien años de soledad", "Gabriel García Márquez", 1967, "978-0307474278")
revista1 = Revista("National Geographic", "National Geographic Society", 2023, "Abril")
tesis1 = Tesis("Impacto del cambio climático en los ecosistemas", "Dr. Ana Pérez", 2022, "Universidad Nacional")

# Mostrar el estado inicial
print("--- Estado Inicial ---")
print(libro1)
print(revista1)
print(tesis1)
print("\n")

# Realizar préstamos
print("--- Realizando préstamos ---")
libro1.prestar()
revista1.prestar()
print("\n")

# Intentar prestar algo no disponible
print("--- Intentando prestar de nuevo ---")
libro1.prestar()
print("\n")

# Mostrar el estado después de préstamos
print("--- Estado después de préstamos ---")
print(libro1)
print(revista1)
print(tesis1)
print("\n")

# Realizar devoluciones
print("--- Realizando devoluciones ---")
libro1.devolver()
print("\n")

# Mostrar el estado después de devoluciones
print("--- Estado después de devoluciones ---")
print(libro1)
print(revista1)
print(tesis1)
```
