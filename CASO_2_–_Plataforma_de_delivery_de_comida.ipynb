```python
class Restaurante:
    def __init__(self, nombre, direccion):
        self.nombre = nombre
        self.direccion = direccion
        self.abierto = False

    def abrir_restaurante(self):
        if not self.abierto:
            self.abierto = True
            print(f"El restaurante {self.nombre} ha abierto sus puertas.")
        else:
            print(f"El restaurante {self.nombre} ya está abierto.")

    def recibir_pedido(self, plato, cantidad):
        if self.abierto:
            print(f"El restaurante {self.nombre} ha recibido un pedido: {cantidad} de {plato}.")
        else:
            print(f"El restaurante {self.nombre} está cerrado y no puede recibir pedidos.")

    def cerrar_restaurante(self):
        if self.abierto:
            self.abierto = False
            print(f"El restaurante {self.nombre} ha cerrado.")
        else:
            print(f"El restaurante {self.nombre} ya está cerrado.")
```

```python
class RestauranteRapido(Restaurante):
    def __init__(self, nombre, direccion, tiempo_entrega_promedio):
        super().__init__(nombre, direccion)
        self.tiempo_entrega_promedio = tiempo_entrega_promedio

    def recibir_pedido(self, plato, cantidad):
        super().recibir_pedido(plato, cantidad)
        if self.abierto:
            print(f"Tiempo de entrega estimado: {self.tiempo_entrega_promedio} minutos.")
```

```python
class RestauranteGourmet(Restaurante):
    def __init__(self, nombre, direccion, chef_ejecutivo):
        super().__init__(nombre, direccion)
        self.chef_ejecutivo = chef_ejecutivo
        self.reservas_llenas = False

    def recibir_pedido(self, plato, cantidad):
        super().recibir_pedido(plato, cantidad)
        if self.abierto and not self.reservas_llenas:
            print(f"Pedido gourmet preparado por el chef {self.chef_ejecutivo}.")
        elif self.abierto and self.reservas_llenas:
            print("Lo sentimos, el restaurante gourmet está lleno para reservas.")

    def llenar_reservas(self):
        self.reservas_llenas = True
        print(f"Las reservas para {self.nombre} están llenas.")
```

```python
class RestauranteVegetariano(Restaurante):
    def __init__(self, nombre, direccion, opciones_veganas=True):
        super().__init__(nombre, direccion)
        self.opciones_veganas = opciones_veganas

    def recibir_pedido(self, plato, cantidad):
        super().recibir_pedido(plato, cantidad)
        if self.abierto:
            print(f"Pedido vegetariano/vegano. Opciones veganas disponibles: {self.opciones_veganas}.")

    def agregar_opcion_vegana(self):
        if not self.opciones_veganas:
            self.opciones_veganas = True
            print(f"Se han agregado opciones veganas al menú de {self.nombre}.")
        else:
            print(f"{self.nombre} ya tiene opciones veganas.")
```

### Demostración de las Clases

```python
# Crear instancias de los restaurantes
rest_generico = Restaurante("La Buena Comida", "Calle Falsa 123")
rest_rapido = RestauranteRapido("Burguer Rápido", "Avenida Siempre Viva 456", 15)
rest_gourmet = RestauranteGourmet("El Delicioso", "Plaza Mayor 789", "Chef Antoine")
rest_vegetariano = RestauranteVegetariano("Verde Fresco", "Bulevar Verde 101", opciones_veganas=True)

# Demostrar funcionalidades
print("\n--- Restaurante Genérico ---")
rest_generico.abrir_restaurante()
rest_generico.recibir_pedido("Pasta", 1)
rest_generico.cerrar_restaurante()
rest_generico.recibir_pedido("Pizza", 2) # Intentar pedir cuando está cerrado

print("\n--- Restaurante Rápido ---")
rest_rapido.abrir_restaurante()
rest_rapido.recibir_pedido("Hamburguesa", 3)

print("\n--- Restaurante Gourmet ---")
rest_gourmet.abrir_restaurante()
rest_gourmet.recibir_pedido("Filet Mignon", 1)
rest_gourmet.llenar_reservas()
rest_gourmet.recibir_pedido("Langosta", 2) # Intentar pedir cuando está lleno

print("\n--- Restaurante Vegetariano ---")
rest_vegetariano.abrir_restaurante()
rest_vegetariano.recibir_pedido("Ensalada", 2)
rest_vegetariano.agregar_opcion_vegana()
```
