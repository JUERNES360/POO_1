```python
class ItemBiblioteca:
    def __init__(self, titulo, codigo): # constructor
        self._titulo = titulo        # atributo protegido
        self._codigo = codigo       # identifica el item
        self._disponible = True      # estado del item

    def prestar(self):
        if self._disponible:
            self._disponible = False
            return f"{self._titulo} ha sido prestado"
        else:
            return f"{self._titulo} no está disponible"

    def devolver(self):
        self._disponible = True
        return f"{self._titulo} ha sido devuelto"

    def descripcion(self):
        return "Item"


item = ItemBiblioteca("Material Multimedia", "I001")

print(item.descripcion())
print(item.prestar())
print(item.prestar())
print(item.devolver())
```
