```python
class Person:

    # Add an __init__ method
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Add a greet method
    def greet(self):
        print("Hello, my name is " + self.name)

# Create an object p1
p1 = Person("JONATHAN", 44)

# Call the greet method
p1.greet()
```
